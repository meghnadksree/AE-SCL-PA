```python
def GetTopNMI(n,CountVectorizer,X,target):
    MI = []
    length = X.shape[1]


    for i in range(length):
        temp=mutual_info_score(X[:, i], target)
        MI.append(temp)
    MIs = sorted(range(len(MI)), key=lambda i: MI[i])[-n:]
    return MIs,MI
```

```python
import numpy as np
from gensim.models import KeyedVectors

def get_bigram_embedding(word1, word2, embedding_model):
    """
    Compute the embedding for a bigram by averaging the embeddings of the individual words.
    """
    if word1 in embedding_model and word2 in embedding_model:
        return np.mean([embedding_model[word1], embedding_model[word2]], axis=0)
    else:
        return None



# Embedding function for multiple pivots
def embed_pivots(pivot_list):
    pivot_embeddings = []  # To store embeddings of all pivot words
    missing_words = []

    for item in pivot_list: # Changed from pivot_list to item to iterate through the tuples within word_pairs
        if len(item) == 3: # Check if the tuple has 3 elements
            kn_word, en_word, synonyms = item
        else:
            kn_word,en_word = item
            synonyms = set()
        # Embed Kannada word
        if kn_word in kn_embeddings:
            kn_embedding = kn_embeddings[kn_word]
        else:
            words = kn_word.split()  # Assuming bigrams are space-separated
            if len(words) == 2:
                    word1, word2 = words
                    bigram_embedding = get_bigram_embedding(word1, word2, kn_embeddings)
                    if bigram_embedding is not None:
                        kn_embeddings.add_vector(kn_word,bigram_embedding)
                    else:
                        missing_words.extend([word1, word2])


        # Embed English translation
        if en_word in en_embeddings:
            en_embedding = en_embeddings[en_word]
        else:
            words = en_word.split()  # Assuming bigrams are space-separated
            if len(words) == 2:
                    word1, word2 = words
                    bigram_embedding = get_bigram_embedding(word1, word2, en_embeddings)
                    if bigram_embedding is not None:
                        en_embeddings.add_vector(en_word,bigram_embedding)
                    else:
                        missing_words.extend([word1, word2])



        # Embed synonyms
        #synonym_embeddings = [en_embeddings[syn] for syn in synonyms if syn in en_embeddings]
        synonym_embeddings = []
        for syn in synonyms:
            if syn in en_embeddings:
                synonym_embeddings.append(en_embeddings[syn])
            else:
                # Check if the synonym is a bigram
                words = syn.split()  # Assuming bigrams are space-separated
                if len(words) == 2:
                    word1, word2 = words
                    bigram_embedding = get_bigram_embedding(word1, word2, en_embeddings)
                    if bigram_embedding is not None:
                        synonym_embeddings.add_vector(en_word,bigram_embedding)
                    else:
                        missing_words.extend([word1, word2])


        # Combine all embeddings (Kannada word + English translation + synonyms)
        all_embeddings = [kn_embedding, en_embedding] + synonym_embeddings
        all_embeddings = [e for e in all_embeddings if e is not None]  # Filter out any None values

        # Average the embeddings if they exist
        if all_embeddings:
            averaged_embedding = np.mean(all_embeddings, axis=0)
            pivot_embeddings.append(averaged_embedding)  # Store the averaged embedding

    print(f"Missing words: {missing_words}")

    return pivot_embeddings

# Embed all the pivot words from the pivot list
pivot_embeddings = embed_pivots(word_pairs)

filename = "/content/drive/My Drive/code/AE-PA-SCL-FT/"+src + "_to_" + dest + "/" +"pivot_mat/"+"pivot_mat_"+src+"_"+dest+"_"+str(pivot_num)+"_"+str(pivot_min_st)+"_"+str(dim)
if not os.path.exists(os.path.dirname(filename)):
      os.makedirs(os.path.dirname(filename))
with open(filename, 'wb') as f:
      pickle.dump(pivot_embeddings, f)

# Output: List of pivot embeddings (one per pivot)
#print(pivot_embeddings)
```
